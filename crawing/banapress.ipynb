```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import os
from pymongo import MongoClient

driver = webdriver.Chrome()
url = 'https://banapresso.com/store'
driver.get(url)

folder_path = '../banapress'

if not os.path.exists(folder_path):
    os.makedirs(folder_path)

def data_info(driver, num):
    image_xpath = f'/html/body/div/div/div/div/article/article/div/div[1]/div[2]/ul/li[{num}]/a/div/img'
    name_xpath = f'/html/body/div/div/div/div/article/article/div/div[1]/div[2]/ul/li[{num}]/a/span/i'
    address_xpath = f'/html/body/div/div/div/div/article/article/div/div[1]/div[2]/ul/li[{num}]/a/span/span'

    image_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, image_xpath))
    )

    name_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, name_xpath))
    )

    address_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, address_xpath))
    )

    name = name_element.text
    address = address_element.text

    image_url = image_element.get_attribute('src')
    image_name = f'../banapress/{name}.jpg'
    image_response = requests.get(image_url, headers={'User-Agent': 'Mozilla/5.0'})

    with open(image_name, 'wb') as image_file:
        image_file.write(image_response.content)

    return name, address, image_name

images = []
names = []
addresses = []

for j in range(3):
    if j == 2:
        for k in range(2, 5):
            if k == 4:
                for m in range(1, 3):
                    name, address, image_name = data_info(driver, m)
                    names.append(name)
                    addresses.append(address)
                    images.append(image_name)
                break       
            for i in range(1, 11):
                name, address, image_name = data_info(driver, i)
                names.append(name)
                addresses.append(address)
                images.append(image_name)

            next_xpath = f'/html/body/div/div/div/div/article/article/div/div[1]/div[3]/ul/li[{k}]/a'
            driver.find_element(By.XPATH, next_xpath).click()
    else:
        for k in range(2, 7):
            for i in range(1, 11):
                name, address, image_name = data_info(driver, i)
                names.append(name)
                addresses.append(address)
                images.append(image_name)

            if k == 6:
                clickBt_xpath = '/html/body/div/div/div/div/article/article/div/div[1]/div[3]/span[2]/a'
                driver.find_element(By.XPATH, clickBt_xpath).click()
                break
                
            next_xpath = f'/html/body/div/div/div/div/article/article/div/div[1]/div[3]/ul/li[{k}]/a'
            driver.find_element(By.XPATH, next_xpath).click()

driver.quit()

url = "mongodb+srv://selfhiam:rVATydvuACjlo9wb@cluster0.z62nxmf.mongodb.net/?retryWrites=true&w=majority"
client = MongoClient(url)
database = client['banapress']
collection = database['stores']

for i in range(len(names)):
    user_insert = {'name': names[i], 'address': addresses[i], 'image': images[i]}
    result = collection.insert_one(user_insert)
    print(f'입력id: {result.inserted_id}')
```
